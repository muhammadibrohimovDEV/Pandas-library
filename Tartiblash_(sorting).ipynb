```python
import pandas as pd
import numpy as np
```

```python
obj = pd.Series(range(4) , index = ['d' , 'a' , 'b' , 'c'])
```

```python
obj
```

```python
obj.sort_index() # .sort_index() Seriesni index lari bo'yicha tartiblaydi
```

```python
obj.sort_values() # .sort_values() Series ni qiymatlari bo'yicha tartiblaydi
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv' , index_col=0)
df.head()
```

```python
df.sort_index() # DataFrame ni indexlari bo'yicha tartiblaydi(axis = 0)
```

```python
df.sort_index(axis = 1) # DataFrame ni ustunlari bo'yicha tartiblaydi (axis = 1 , inplace = True)
```

```python
df.sort_index(axis = 1 , inplace = True , ascending = False) # ascending = False bu teskari tartiblah (yuqoridan , pastga)
```

```python
df
```

```python
df.sort_index(ascending = False)
```

```python
df.sort_values(by = 'price' , ascending = False , inplace = True)
df
```

```python
df.sort_values(by = ['price' , 'mileage'])
```
